```python
import pandas as pd
import numpy as np
import matplotlib as mpl
from faker import Faker
```

```python
np.random.seed(0)

fake = Faker()
date_range = pd.date_range(start='2022-01-01', end='2023-12-31', freq='D')
df = pd.DataFrame({
    'date': pd.to_datetime(np.random.choice(date_range, 10)),
    'price': np.random.uniform(10, 100, 10),
    'rate': np.random.uniform(10, 100, 10),
    'volume': np.random.uniform(10, 100, 10),
    'ID': np.random.choice(['A', 'B', 'C'], 10)
    })
```

```python
df
```

```python
df_style = (
    df.style
    .format(precision=4, thousands=".", decimal=",")
    .highlight_max(["price"], color="blue")
    .highlight_min(["price"], color="red")
    .background_gradient(subset="rate", cmap="Greens")
    .bar('volume', color='lightblue', )  # align='zero'
    )
```

```python
df_style
```

```python
from reportlab.lib.pagesizes import letter
from reportlab.platypus import SimpleDocTemplate, Table, TableStyle
from reportlab.lib import colors
```

```python
html = df_style.to_html()
```

```python
pdf_file = "styled_dataframe.pdf"
doc = SimpleDocTemplate(pdf_file, pagesize=letter)
table = Table(pd.read_html(html)[0].values)
table.setStyle(TableStyle([('BACKGROUND', (0, 0), (-1, 0), colors.grey),
                           ('TEXTCOLOR', (0, 0), (-1, 0), colors.whitesmoke),
                           ('ALIGN', (0, 0), (-1, -1), 'CENTER'),
                           ('FONTNAME', (0, 0), (-1, 0), 'Helvetica-Bold'),
                           ('BOTTOMPADDING', (0, 0), (-1, 0), 12),
                           ('BACKGROUND', (0, 1), (-1, -1), colors.beige)]))

doc.build([table])
```
